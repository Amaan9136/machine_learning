```python
import matplotlib.pyplot as plt 
import pandas
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Load the Wine Quality Red dataset
file = pandas.read_csv("../../datasets/iris.csv")

# Separate features and target
X=file.iloc[:,:-1].values
y=file.iloc[:,-1].values

# Split the dataset into an 80-20 training-test set
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
# regressor model creation (multiple regression)
regressor = LinearRegression()

# train model with multiple regression
regressor.fit(X_train, y_train)

# predict test data 
y_pred = regressor.predict(X_test)
```

```python
# Concatenate X_test, y_test (actual), and y_pred (predicted) using np concatenate
data_comparison = np.concatenate((X_test, y_test.reshape(-1, 1), y_pred.reshape(-1, 1)), axis=1)

# Displaying the concatenated data with headings
headings = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width', 'Actual', 'Predicted']
print('{:<15} {:<15} {:<15} {:<15} {:<15} {:<15}'.format(*headings))

for row in data_comparison:
    print('{:<15.1f} {:<15.1f} {:<15.1f} {:<15.1f} {:<15.1f} {:<15.1f}'.format(*row))
```

```python
# Extract individual test columns for plotting 

sepal_length_X = X_test[:, 0]
sepal_width_X = X_test[:, 1]
petal_length_X = X_test[:, 2]
petal_width_X = X_test[:, 3]
```

```python
# compare the individual columns with y predicted values by plotting graph

# Plotting predicted vs actual for each feature
plt.figure(figsize=(15, 5))

# Sepal Length
plt.subplot(1, 4, 1)
plt.scatter(sepal_length_X, y_pred, color='blue', label='Predicted')
plt.scatter(sepal_length_X, y_test, color='red', label='Actual', marker='x')
plt.xlabel('Sepal Length')
plt.ylabel('Values')
plt.title('Sepal Length: Predicted vs Actual')
plt.legend()

# Sepal Width
plt.subplot(1, 4, 2)
plt.scatter(sepal_width_X, y_pred, color='blue', label='Predicted')
plt.scatter(sepal_width_X, y_test, color='red', label='Actual', marker='x')
plt.xlabel('Sepal Width')
plt.ylabel('Values')
plt.title('Sepal Width: Predicted vs Actual')
plt.legend()

# Petal Length
plt.subplot(1, 4, 3)
plt.scatter(petal_length_X, y_pred, color='blue', label='Predicted')
plt.scatter(petal_length_X, y_test, color='red', label='Actual', marker='x')
plt.xlabel('Petal Length')
plt.ylabel('Values')
plt.title('Petal Length: Predicted vs Actual')
plt.legend()

# Petal Width
plt.subplot(1, 4, 4)
plt.scatter(petal_width_X, y_pred, color='blue', label='Predicted')
plt.scatter(petal_width_X, y_test, color='red', label='Actual', marker='x')
plt.xlabel('Petal Width')
plt.ylabel('Values')
plt.title('Petal Width: Predicted vs Actual')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# custom predictions

print(regressor.predict([[6.4,2.4,4.2,1.6]]))
# given test value the output will be predicted value
```

```python
# score 

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2) Score: {r2}")
print(f"Mean Absolute Error (MAE): {mae}")
```
