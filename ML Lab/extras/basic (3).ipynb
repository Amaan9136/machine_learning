```python
#LAB PROGRAM 1


import matplotlib.pyplot as plt
hours_studied=[10,9,2,15,10,16,11,16]
exam_scores=[95,80,10,50,45,98,38,93]
plt.figure(figsize=(8,8))
plt.plot(hours_studied,exam_scores,marker='*',color='red',linestyle='-',markersize=10)
plt.xlabel("Number of hours spent studying")
plt.ylabel("score in final exam (0-100)")
plt.title("Effect of study on exam performance")
plt.grid(True)
plt.show()
```
