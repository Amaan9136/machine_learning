# GENERATE GRAPHS AND METRICS AND STORE

## HISTOGRAM GENERATOR 1

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd  # Import pandas for handling CSV files

# Generate 100 random API request durations (in seconds) between 0 and 22 seconds
requests_duration = np.random.uniform(0, 22, 100)

# Calculate metrics
mean_duration = np.mean(requests_duration)
median_duration = np.median(requests_duration)
std_dev_duration = np.std(requests_duration)

# Create a pandas DataFrame to store the request durations
data = {'Request Duration (sec)': requests_duration}
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('STORE_DATA.csv', index=False)

# Plotting the histogram
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the histogram for request durations with nicer color and style
ax.hist(requests_duration, bins=10, color='teal', edgecolor='black', alpha=0.7)

# Highlight the RSI region between 10 and 20 seconds
ax.axvspan(10, 20, color='yellow', alpha=0.3, label="RSI (10-20 seconds)")  # Highlight RSI region

# Add the mean, median, and standard deviation as lines on the graph
ax.axvline(mean_duration, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_duration:.2f} sec')
ax.axvline(median_duration, color='green', linestyle='-', linewidth=2, label=f'Median: {median_duration:.2f} sec')
ax.axvline(mean_duration + std_dev_duration, color='orange', linestyle='-', linewidth=2, label=f'+1 Std Dev: {mean_duration + std_dev_duration:.2f} sec')
ax.axvline(mean_duration - std_dev_duration, color='orange', linestyle='-', linewidth=2, label=f'-1 Std Dev: {mean_duration - std_dev_duration:.2f} sec')

# Highlight all 100 requests with blue dots
for duration in requests_duration:
    ax.plot(duration, 0, 'bo', markersize=5)  # 'bo' creates blue dots for each request

# Add labels, title, and grid
ax.set_xlabel('Request Duration (seconds)', fontsize=12)
ax.set_ylabel('Number of Requests', fontsize=12)
ax.set_title('Request Sent to Train Model (100 Requests) and Time Taken to Train (30s)', fontsize=14)
ax.grid(True, linestyle='--', alpha=0.5)

# Set a texture on the background for better visualization
ax.set_facecolor('whitesmoke')  # Set a soft background color

# Show the legend
ax.legend()

# Show the plot
plt.show()
```

## PRE TRAINED MODEL REQUEST VS TIME 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd  # Import pandas for handling CSV files

# Generate 100 random processing times (in seconds) between 0 and 60 seconds
pretrained_model_times = np.random.uniform(0, 60, 100)

# Calculate metrics
mean_time = np.mean(pretrained_model_times)
median_time = np.median(pretrained_model_times)
std_dev_time = np.std(pretrained_model_times)

# Create a pandas DataFrame to store the processing times
data = {'Processing Time (sec)': pretrained_model_times}
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('STORE_DATA.csv', index=False)

# Plotting the histogram
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the histogram for request durations (time taken by the pretrained model)
ax.hist(pretrained_model_times, bins=10, color='teal', edgecolor='black', alpha=0.7)

# Highlight the region between 20 and 40 seconds as RSI
ax.axvspan(20, 40, color='yellow', alpha=0.3, label="RSI (20-40 seconds)")  # Highlight RSI region

# Add the mean, median, and standard deviation as lines on the graph
ax.axvline(mean_time, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_time:.2f} sec')
ax.axvline(median_time, color='green', linestyle='-', linewidth=2, label=f'Median: {median_time:.2f} sec')
ax.axvline(mean_time + std_dev_time, color='orange', linestyle='-', linewidth=2, label=f'+1 Std Dev: {mean_time + std_dev_time:.2f} sec')
ax.axvline(mean_time - std_dev_time, color='orange', linestyle='-', linewidth=2, label=f'-1 Std Dev: {mean_time - std_dev_time:.2f} sec')

# Add labels, title, and grid
ax.set_xlabel('Processing Time (seconds)', fontsize=12)
ax.set_ylabel('Number of Requests', fontsize=12)
ax.set_title('Pretrained Model Time Taken (60s) to Process Requests (100 Requests)', fontsize=14)
ax.grid(True, linestyle='--', alpha=0.5)

# Set a texture on the background for better visualization
ax.set_facecolor('whitesmoke')  # Set a soft background color

# Show the legend
ax.legend()

# Show the plot
plt.show()
```
