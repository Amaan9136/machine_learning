```python
import matplotlib.pyplot as plt
import pandas
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Load the Wine Quality Red dataset
file = pandas.read_csv("studied_min_cgpa.csv")

# Separate features and target
X=file.iloc[:,:-1].values
y=file.iloc[:,-1].values
# Split the dataset into an 80-20 training-test set
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
# regressor model creation
regressor = LinearRegression()

# train model
regressor.fit(X_train, y_train)

# predict test data
y_pred = regressor.predict(X_test)
```

```python
# plot the train data's points
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.xlabel = 'Studied mins'
plt.ylabel = 'CGPA'
plt.title = 'Studied mins v/s CGPA'
plt.show()
```

```python
# plot the test data's points
plt.scatter(X_test, y_pred, color='blue', label='Predicted')
plt.scatter(X_test, y_test, color='red', label='Actual', marker='x')
# plot the line for x-test's predicted data which is y-test-predicted
plt.plot(X_test,y_pred,color='green')
plt.xlabel = 'Studied mins'
plt.ylabel = 'CGPA'
plt.title = 'Studied mins v/s CGPA'
plt.legend()
plt.show()
```
